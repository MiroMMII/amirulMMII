```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 1. SIMULASI DATA (Data penggunaan AWS anda selama 7 hari lepas)
# Hari ke-x vs Kos dalam USD
data_penggunaan = {
    'Hari': [1, 2, 3, 4, 5, 6, 7],
    'Kos_USD': [1.2, 2.5, 3.8, 5.0, 6.2, 7.5, 8.9]
}

df = pd.DataFrame(data_penggunaan)
X = df[['Hari']] # Feature
y = df['Kos_USD'] # Target

# 2. MELATIH AI (Machine Learning)
model = LinearRegression()
model.fit(X, y)

# 3. RAMALAN UNTUK HARI KE-30 (Hujung Bulan)
hari_ke_30 = np.array([[30]])
ramalan_30_hari = model.predict(hari_ke_30)

# 4. LOGIK KEPUTUSAN (Decision Making)
had_bajet = 35.00 # USD
status_bajet = "SELAMAT" if ramalan_30_hari[0] <= had_bajet else "AMARAN: AKAN LEBIH BAJET!"

# 5. PAPARKAN HASIL
print(f"--- LAPORAN AI AWS COST PREDICTOR ---")
print(f"Kos hari ini (Hari 7): ${df['Kos_USD'].iloc[-1]}")
print(f"Ramalan kos pada hari ke-30: ${ramalan_30_hari[0]:.2f}")
print(f"Had Bajet: ${had_bajet}")
print(f"Status: {status_bajet}")

# 6. VISUALISASI (Supaya nampak pro dalam portfolio)
plt.scatter(X, y, color='blue', label='Data Sebenar')
plt.plot(X, model.predict(X), color='red', label='Trend AI')
plt.title('Ramalan Kos AWS')
plt.xlabel('Hari')
plt.ylabel('Kos (USD)')
plt.legend()
plt.show()
```
